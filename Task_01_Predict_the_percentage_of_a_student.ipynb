# Predict the percentage of a student based on the no. of study hours.

## Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
# Load the data
raw_data = pd.read_csv('task_01_data.csv')
raw_data.head()
```

```python
# plot out data
sns.scatterplot(data=raw_data, x='Hours', y='Scores')
```

```python
# Declare the dependent and independent variables
x = raw_data['Hours']
y = raw_data['Scores']
x.shape, y.shape
```

```python
Xmatrix = x.values.reshape(-1,1)
Xmatrix.shape
```

## Regression

```python
# Creating a regression Object
reg = LinearRegression()
```

```python
reg.fit(Xmatrix,y)
```

```python
# predict the same input to get the regression line
y_pred = reg.predict(Xmatrix)
```

```python
# plot regression line
sns.scatterplot(data=raw_data, x='Hours', y='Scores')
sns.lineplot(x=raw_data['Hours'], y=y_pred)
```

## Predict new data

```python
newx = np.array([9.25]).reshape(-1,1)
reg.predict(newx)
```

### So our model specifies that if the student attends 9.25 hours, the expected score will be around 92.9.
